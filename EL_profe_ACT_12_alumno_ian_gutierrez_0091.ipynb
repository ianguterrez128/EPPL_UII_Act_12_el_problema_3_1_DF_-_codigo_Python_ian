## Las variables
1. CP --> costo del producto
2. CD --> cantidad de dinero
3. Cambio

## Formula
$Cambio = CD - CP$

## Diagrama de flujo lucidchart
 ![texto del enlace](https://raw.githubusercontent.com/ianguterrez128/EPPL_UII_Act_12_el_problema_3_1_DF_-_codigo_Python_ian/refs/heads/main/coco1.png)

```python
# para cuando es verdadero (si)
# Calculando el camnbio usando if else
# Datos de entrada CD , CP
# mostrar cambio si es falso
# si es verdadero mostrar no alcanza
CD = float(input("Ingrese la cantidad de dinero: "))
CP = float(input("Ingrese el costo del producto: "))
if CP >= CD:
  print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por ian gutierrez NL = 22 ")
```

```python
# para cuando es falso (no)
# Calculando el camnbio usando if else
# Datos de entrada CD , CP
# mostrar cambio si es falso
# si es verdadero mostrar no alcanza
CD = float(input("Ingrese la cantidad de dinero: "))
CP = float(input("Ingrese el costo del producto: "))
if CP >= CD:
  print("No alcanza")
else:
  Cambio = CD - CP
  print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por ian gutierrez NL = 22 ")
```
